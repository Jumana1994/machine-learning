```python
x=["jum","saju","arun","amal","sarun"]
y=[5,4,7,6,9]
```

```python
from matplotlib import pyplot as plt
```

```python
plt.plot(x,y)
```

```python
plt.plot(x,y,marker="o",color="g",linestyle="-.",lw=3)
plt.title("student marks")
plt.xlabel("names")
plt.ylabel("marks")
plt.grid()
```

```python
a=['apple','banana','orange','kivi','watermelon']
b=[55,46,78,35,49]
c=[77,88,44,95,67]
```

```python
from matplotlib import pyplot as plt
```

```python
plt.plot(a,b,marker="*",color="b",linestyle="-.",lw=1,label="quarterly marks")
plt.plot(a,c,marker="o",color="r",linestyle="--",lw=1,label="halfyearly marks")
plt.legend(loc="upper right")
plt.title("student marks")
plt.xlabel("names")
plt.ylabel("marks")
```

```python
plt.bar(a,b,color="b",width=.5,label="quarterly marks",edgecolor="y")
plt.bar(a,c,color="r",width=.3,label="halfyearly marks",edgecolor="y")
plt.title("student marks")
plt.xlabel("names")
plt.ylabel("marks")
plt.legend()
# plt.xticks(ticks=b,labels=a)
```

```python
plt.barh(a,b)
```

```python
plt.scatter(a,b,marker="*",color="b",linestyle="-.",lw=1,label="quarterly marks")
plt.title("student marks")
plt.xlabel("names")
plt.ylabel("marks")
```
