```python
import sklearn
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv("carprices.csv")
```

```python
data
```

```python
data.isnull().any(axis=0)
```

```python
dummy=pd.get_dummies(data["Car Model"])
dummy
```

```python
data1=pd.concat([data,dummy],axis=1)
data1
```

```python
x=data1[["Audi A5","BMW X5","Mercedez Benz C class","Mileage","Age(yrs)"]]
y=data1[["Sell Price($)"]]
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model=LinearRegression()
```

```python
model.fit(x,y)
```

```python
model.predict([[1,0,0,52000,5]])
```

```python
model.score(x,y)
```
