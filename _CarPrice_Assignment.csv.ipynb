```python
import sklearn
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv("CarPrice_Assignment.csv")
data
```

```python
d1=pd.get_dummies(data["CarName"])
d2=pd.get_dummies(data["fueltype"])
d3=pd.get_dummies(data["doornumber"])
d4=pd.get_dummies(data["carbody"])
d5=pd.get_dummies(data["enginetype"])
data=pd.concat([data,d1,d2,d3,d4,d5],axis=1)
data

```

```python
data.drop(["CarName","fueltype","doornumber","carbody","enginetype"],axis=1,inplace=True)
data
```

```python
data.drop(["car_ID","symboling","aspiration","drivewheel","enginelocation","wheelbase","carlength","carwidth","carheight","curbweight","cylindernumber","enginesize","fuelsystem","boreratio","stroke","compressionratio","horsepower","peakrpm","citympg","highwaympg"],axis=1,inplace=True)
```

```python
data
```

```python
x=data.drop(["price"],axis=1)
y=data["price"]
```
