```python
import sklearn
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv("salaries.csv")
```

```python
data
```

```python
d1=pd.get_dummies(data["company"])
d2=pd.get_dummies(data["degree"])
d3=pd.get_dummies(data["job"])
data=pd.concat([data,d1,d2,d3],axis=1)
data
```

```python
data.drop(["company","job","degree"],inplace=True,axis=1)
```

```python
data
```

```python
x=data.drop(["salary_more_then_100k"],axis=1)
y=data["salary_more_then_100k"]
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model=LogisticRegression()
```

```python
model.fit(x,y)
```

```python
model.score(x,y)
```
